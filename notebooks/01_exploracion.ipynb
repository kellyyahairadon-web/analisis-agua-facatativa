```python
import pandas as pd
```

```python
datos = pd.read_csv("../datos/datos_experimentales.csv")

datos.head()
```

```python
datos.columns
```

```python
datos.describe()
```

```python
print("Promedio pH:", datos["ph"].mean())
print("Promedio oxígeno:", datos["oxigeno"].mean())
print("Promedio temperatura:", datos["temperatura"].mean())
```

```python
print("pH mínimo:", datos["ph"].min())
print("pH máximo:", datos["ph"].max())

print("Oxígeno mínimo:", datos["oxigeno"].min())
print("Oxígeno máximo:", datos["oxigeno"].max())

print("Temperatura mínima:", datos["temperatura"].min())
print("Temperatura máxima:", datos["temperatura"].max())
```

```python
datos[(datos["ph"] < 0) | (datos["ph"] > 14)]
```

```python
datos[(datos["temperatura"] < 0)]
```

```python
import matplotlib.pyplot as plt

plt.scatter(datos["temperatura"], datos["oxigeno"])

plt.xlabel("Temperatura")
plt.ylabel("Oxígeno")

plt.title("Relación entre temperatura y oxígeno")

plt.show()
```

## Análisis de resultados

Se realizó una exploración inicial de los datos relacionados con la calidad del agua, considerando variables como pH, oxígeno disuelto, temperatura, conductividad y dureza.

Las estadísticas descriptivas permitieron identificar el comportamiento general del sistema y verificar que los datos no presentan anomalías importantes, ya que no se encontraron valores de pH fuera del rango normal ni temperaturas negativas.

La gráfica de dispersión entre temperatura y oxígeno mostró una relación inversa entre ambas variables. Se observó que al aumentar la temperatura, la concentración de oxígeno disuelto disminuye.

Este comportamiento coincide con la teoría de calidad del agua, donde temperaturas más altas reducen la capacidad del agua para retener oxígeno.

## Conclusión

El análisis exploratorio permitió comprender el comportamiento de las principales variables relacionadas con la calidad del agua. Además, se evidenció una relación entre la temperatura y el oxígeno disuelto, coherente con lo esperado teóricamente.

El uso de herramientas de análisis de datos en Python facilitó la organización, exploración e interpretación de la información experimental.
