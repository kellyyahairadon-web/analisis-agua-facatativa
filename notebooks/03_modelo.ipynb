# Reto 3 – Funciones y análisis

En este reto se crean funciones para evaluar la calidad del agua y construir un índice de contaminación utilizando Python y pandas.

```python
import pandas as pd

df = pd.read_csv("../datos/datos_experimentales.csv")

df.head()
```

```python
def evaluar_calidad(oxigeno):

    if oxigeno < 5:
        return "Mala"

    elif oxigeno <= 7:
        return "Regular"

    else:
        return "Buena"
```

```python
df["calidad_agua"] = df["oxigeno"].apply(evaluar_calidad)

df.head()
```

```python
def indice_contaminacion(ph, oxigeno):

    return abs(7 - ph) + (10 - oxigeno)
```

```python
df["indice_contaminacion"] = df.apply(
    lambda fila: indice_contaminacion(
        fila["ph"],
        fila["oxigeno"]
    ),
    axis=1
)

df.head()
```

```python
df["indice_contaminacion"].describe()
```

```python
df["calidad_agua"].value_counts()
```

## Interpretación de resultados

Las funciones desarrolladas permitieron clasificar la calidad del agua y construir un índice de contaminación basado en el pH y el oxígeno disuelto.

Los resultados muestran que valores bajos de oxígeno y desviaciones del pH neutro aumentan el índice de contaminación, indicando condiciones menos favorables para el ecosistema acuático.

El análisis evidencia la importancia de aplicar programación y lógica matemática en la evaluación de variables ambientales.
